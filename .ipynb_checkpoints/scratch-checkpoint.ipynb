```python
from kinematics_likelihood import KinematicLikelihood
```

```python
likelihood = KinematicLikelihood(software='galkin', anisotropy_model='constant', 
                                 aperture='single_slit', is_spherical=True
                                )
```

```python
import numpy as np

a = np.random.multivariate_normal(likelihood.lens_model_posterior_mean, 
                     cov=likelihood.lens_model_posterior_covariance,
                                  size=32
                                 )

a.shape
```

```python
ls = []
for b in a:
    ls.append(likelihood.lens_model_likelihood(b))
```

```python
import seaborn as sns

sns.kdeplot(np.exp(ls))
```
