## check anisotropy profile

```python
from copy import deepcopy

np.random.seed(2)

self = dyn_model
r_eff_uncertainty_factor = np.random.normal(loc=1., scale=0.05)

r_eff = 1.85 * r_eff_uncertainty_factor #self.get_r_eff(i)
# Jeffrey's prior for a_ani

ani_param = 10**np.random.uniform(np.log10(0.5),
                              np.log10(5)) * r_eff

kwargs_light = self.get_double_sersic_kwargs(is_shperical=True)
e11, e12, e21, e22 = self.get_e_parameters()

surf_lum, sigma_lum, qobs_lum = self.get_light_profile_mge(
    deepcopy(kwargs_light), e11, e12, e21, e22)

surf_pot, sigma_pot, qobs_pot = self.get_mass_mge(sample_index=0)

bs = self.get_bs(ani_param, surf_lum, sigma_lum,
                 model='Osipkov-Merritt', plot=True
                 )
```

## constant anisotropy

```python
from jampy.jam_axi_proj import jam_axi_proj
from jampy.jam_sph_rms import jam_sph_rms


# check v_rms
Rs_galkin = np.linspace(0, 2, 100)
Rs = Rs_galkin[:-1] + (Rs_galkin[1] - Rs_galkin[0])/2.

beta = 0.3

v_rms_galkin = self.compute_galkin_v_rms_radial_profile(Rs_galkin,
                                                        anisotropy_model='constant',
                                                        ani_param=beta
                                                       )

#print(v_rms_galkin)

bs = np.ones_like(sigma_lum) * beta

v_rms_axi = jam_axi_proj(surf_lum, sigma_lum, np.ones_like(sigma_lum), 
                         surf_pot, sigma_pot, np.ones_like(sigma_pot),
                         inc=90, mbh=0, 
                         distance=self.lens_cosmo.dd, 
                         xbin=Rs, ybin=Rs*0, 
                         beta=bs, 
                         align='sph').model

v_rms_sph = jam_sph_rms(surf_lum, sigma_lum, 
                        surf_pot, sigma_pot,
                        mbh=0, 
                        distance=self.lens_cosmo.dd, 
                        rad=Rs,
                        beta=bs, 
                        #align='sph'
                       )[0]

plt.plot(Rs, v_rms_galkin, label='Galkin numeric')
plt.plot(Rs, v_rms_axi, label='jampy (axi)')
plt.plot(Rs, v_rms_sph, label='jampy (sph)')

plt.legend()
plt.xlabel(r'$R$ (arcsec)')
plt.ylabel(r'$v_{\rm rms}$ (km s$^{-1}$)');
plt.title('All lines should match!')
plt.show()


plt.plot(Rs, v_rms_galkin/v_rms_axi, label='Galkin numeric')
plt.plot(Rs, v_rms_sph/v_rms_axi, label='jampy (sph)')

plt.legend()
plt.xlabel(r'$R$ (arcsec)')
plt.ylabel(r'$v_{\rm rms}/v_{\rm jampy}$');
# plt.title('All lines should match!')
plt.show()
```
