```python
delta_pa = np.array([
    2.6038461538461535,
    1.603846153846154,
    1.1,
    1.103846153846154,
    1.5038461538461538,
    1.3038461538461537,
    2.1038461538461535,
    2.6038461538461535,
    2.9038461538461537,
    3.407692307692307,
])

delta_pa_high = np.array([
    3.2076923076923074,
    1.9038461538461555,
    1.353846153846154,
    1.4038461538461484,
    1.8038461538461554,
    1.611538461538463,
    2.5038461538461547,
    3.0076923076923077,
    3.3038461538461537,
    3.853846153846153
])

delta_pa_uncertainty = (delta_pa_high - delta_pa)
```

```python
bar_points = np.linspace(-90, 90, 20)
bar_points_half = ((bar_points[1:] + bar_points[:-1]) / 2.)[-10:]

bar_points_half[0] = 0

bar_points_half = 90 - bar_points_half

bar_points_half
```

```python
def func(params):
    """
    """
    amp1, sig1, amp2, sig2 = params
    
    density = amp1 * np.exp(-bar_points_half**2/2/sig1**2) \
            + amp2 * np.exp(-(90-bar_points_half)**2/2/sig2**2)
    
    return 0.5 * np.sum((density  - delta_pa)**2 / delta_pa_uncertainty**2)

def evaluate_func(x, params):
    """
    """
    amp1, sig1, amp2, sig2 = params
    
    density = amp1 * np.exp(-x**2/2./sig1**2) \
            + amp2 * np.exp(-(90-x)**2/2./sig2**2)
    
    return density
```

```python
from scipy.optimize import minimize 

res = minimize(func, x0=[2.5, 15, 3.2, 35])

res
```

```python
from dynamical_model import DynamicalModel

dyn_model = DynamicalModel('powerlaw')
```

```python
def get_profiles(xs, ys, shape, moment='zz'):
    """
    """
    if shape == 'prolate':
        params = prolate_params
    else:
        params = oblate_params
        
    theta_e, gamma, q, pa, D_dt, inclination, lamda, ani_param = params
    lens_params = [theta_e, gamma, q]
        
    return dyn_model.compute_jampy_v_rms_model(lens_params, ani_param=ani_param,
                                               pa=pa, inclination=inclination,
                                        anisotropy_model='constant',
                                        moment=moment,
                                        x_points=xs, y_points=ys,
                                        shape=shape
                                       ) * np.sqrt(lamda)
```
