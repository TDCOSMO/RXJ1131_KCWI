```python
import numpy as np
import matplotlib.pyplot as plt

from lenstronomy.LightModel.light_model import LightModel
from lenstronomy.Util.param_util import phi_q2_ellipticity
```

```python
light_model = LightModel(['SERSIC_ELLIPSE', 'SERSIC_ELLIPSE'])

e11, e12 = phi_q2_ellipticity(45, 0.878) # (90 - 121.6)*
e21, e22 = phi_q2_ellipticity(45, 0.849) # (90 - 121.6)*

kwargs_light = [
            {'R_sersic': 2.49*np.sqrt(0.878), 'n_sersic': 0.93, 'amp': 0.091, 
             'center_x': 0, 'center_y': 0,
             'e1': e11, 'e2': e12
            },
            {'R_sersic': 0.362*np.sqrt(0.849), 'n_sersic': 1.59, 'amp': 0.89, 
             'center_x': 0, 'center_y': 0,
             'e1': e11, 'e2': e12
            },
        ]
```

```python
w = 1.85 * 50 / np.sqrt(2)

x, y = np.meshgrid(
        np.linspace(-w, w, 500), # x-axis points to negative RA
        np.linspace(-w, w, 500),
    )

surface_brightness = light_model.surface_brightness(x, y, kwargs_light)

plt.matshow(np.log10(surface_brightness), cmap='cubehelix', origin='lower')
plt.show()
```

```python
from mgefit.find_galaxy import find_galaxy
from mgefit.mge_fit_sectors_regularized import mge_fit_sectors_regularized
from mgefit.sectors_photometry import sectors_photometry


img = surface_brightness

scale = np.abs(x[0, 1] - x[0, 0]) #0.04  # arcsec/pixel
minlevel = 1e-49 # counts/pixel
ngauss = 20

sigmapsf = 0     # In pixels
normpsf = 1  # total(normpsf)=1


# Here we use FIND_GALAXY directly inside the procedure. Usually you may want
# to experiment with different values of the FRACTION keyword, before adopting
# given values of Eps, Ang, Xc, Yc.
plt.clf()
f = find_galaxy(img, fraction=0.3, plot=1)
plt.show()

# Perform galaxy photometry
plt.clf()
s = sectors_photometry(img, f.eps, f.theta, f.xpeak, f.ypeak,
                       minlevel=minlevel, plot=1)
plt.show()

# Do the actual MGE fit
# *********************** IMPORTANT ***********************************
# For the final publication-quality MGE fit one should include the line
# "from mge_fit_sectors_regularized import mge_fit_sectors_regularized"
# at the top of this file, rename mge_fit_sectors() into
# mge_fit_sectors_regularized() and re-run the procedure.
# See the documentation of mge_fit_sectors_regularized for details.
# *********************************************************************
plt.clf()
m = mge_fit_sectors_regularized(s.radius, s.angle, s.counts, f.eps,
                    ngauss=ngauss, sigmapsf=sigmapsf, normpsf=normpsf,
                    scale=scale, plot=1, bulge_disk=0, linear=0)
plt.show()
```

```python
from mgefit.mge_print_contours import mge_print_contours


# Show contour plots of the results
plt.clf()
plt.subplot(121)
mge_print_contours(img.clip(minlevel), f.theta, f.xpeak, f.ypeak, m.sol, scale=scale,
                   binning=3, sigmapsf=sigmapsf, normpsf=normpsf, magrange=80)

# Extract the central part of the image to plot at high resolution.
# The MGE is centered to fractional pixel accuracy to ease visual comparison.

n = 100
img_clipped = img[f.xpeak-n:f.xpeak+n, f.ypeak-n:f.ypeak+n]
xc, yc = n - f.xpeak + f.xmed, n - f.ypeak + f.ymed
plt.subplot(122)
#mge_print_contours(img_clipped.clip(minlevel), f.theta, xc, yc, m.sol, binning=0.2,
#                  sigmapsf=sigmapsf, normpsf=normpsf, scale=scale)
mge_print_contours(img.clip(minlevel), f.theta, f.xpeak, f.ypeak, m.sol, scale=scale,
                   binning=1, sigmapsf=sigmapsf, normpsf=normpsf, magrange=50)
# plt.xlim(13, 18)
# plt.ylim(13, 18)
```

```python
sigmas = m.sol[1, :] * np.sqrt(1 - f.eps) * scale
amps = m.sol[0, :] * scale**2
```

```python
sigmas, amps
```

```python
rs = np.logspace(-1, 1, 100)

mge_profile = np.zeros_like(rs)

for a, s in zip(amps, sigmas):
    mge_profile += a / np.sqrt(2*np.pi * s**2)**2 * np.exp(-rs**2 / 2 / s**2 )
    
plt.loglog(rs, light_model.surface_brightness(rs, rs*0, kwargs_light), label='True')
plt.loglog(rs, mge_profile, label='MGE')
plt.legend()
plt.show()
```

## Fit Hernquist profile to radial profile

```python
double_sersic_model = LightModel(['SERSIC', 'SERSIC'])

rs = np.logspace(-1.6, 2, 300)
kwargs_double_sersic = [
        {'R_sersic': 2.49*np.sqrt(0.878), 'n_sersic': 0.93, 'amp': 0.091, 'center_x': 0, 'center_y': 0},
        {'R_sersic': 0.362*np.sqrt(0.849), 'n_sersic': 1.59, 'amp': 0.89, 'center_x': 0, 'center_y': 0},
    ]
    
double_sersic_radial_profile = double_sersic_model.surface_brightness(rs, 0*rs, kwargs_double_sersic)

hernquist_model = LightModel(['HERNQUIST'])

def cost_function(params, plot=False):
    """
    """
    kwargs_hernquist = [{'Rs': params[0], 'amp': params[1], 'center_x': 0, 'center_y': 0}]
    
    hernquist_profile = hernquist_model.surface_brightness(rs, 0*rs, kwargs_hernquist)
    
    if plot:
        plt.loglog(rs, double_sersic_radial_profile, label='double Sersic')
        plt.loglog(rs, hernquist_profile, label='Hernquist')
        plt.legend()
        plt.show()
        
    noise = np.sqrt(double_sersic_radial_profile * 2.5)/double_sersic_radial_profile
    
    return np.sum((double_sersic_radial_profile - hernquist_profile)**2/noise**2)
```

```python
from scipy.optimize import minimize 

minimize(cost_function, x0=[1.85, 1])
```
